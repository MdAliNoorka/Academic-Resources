## Setup

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Economy_of_US.csv')
df.head()
```

## A simple splot

```python
plt.plot(df['Year'], df['GDP_Growth'])
```

```python
plt.plot(df['Year'], df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
plt.show()
```

### Change markers

```python
plt.plot(df['Year'], df['GDP_Growth'], 'o')
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
plt.show()
```

```python
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 
           'H', 'h', 'v', '^', '<', '>', '1', '2', '3', '4', '|', '_']

for m in markers:
    print(m)
    plt.plot(df['Year'], df['GDP_Growth'], m)
    plt.xlabel('Year')
    plt.ylabel('GDP GRowth')
    plt.title('Economy of US')
    plt.show()
```

### Change line style

```python
line = ['-',':','--','-.']
for l in line:
    print(l)
    l = 'o' + l + 'r'
    plt.plot(df['Year'], df['GDP_Growth'], l)
    plt.xlabel('Year')
    plt.ylabel('GDP GRowth')
    plt.title('Economy of US')
    plt.show()
```

### Change color

```python
color = ['r','g','b','c','m','y','k','w']
for c in color:
    print(c)
    c = 'o:' + c
    plt.plot(df['Year'], df['GDP_Growth'], c)
    plt.xlabel('Year')
    plt.ylabel('GDP GRowth')
    plt.title('Economy of US')
    plt.show()
```

### Change marker size

```python
plt.plot(df['Year'], df['GDP_Growth'], marker = 'o', ms = 5)
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
plt.show()
```

## Multiplots

```python
plt.plot(df['Year'], df['GDP_Growth'])
plt.plot(df['Year'], df['Inflation'])
plt.plot(df['Year'], df['Unemployment'])
plt.xlabel('Year')
plt.title('Economy of US')
```

### Add legend

```python
plt.plot(df['Year'], df['GDP_Growth'], label = 'GDP_Growth')
plt.plot(df['Year'], df['Inflation'], label = 'Inflation' )
plt.plot(df['Year'], df['Unemployment'], label = 'Unemployment')
plt.xlabel('Year')
plt.ylabel('Economy')
plt.grid()
plt.legend()
```

### Change arrangement

```python
plt.subplot(3, 1, 1)
plt.plot(df['Year'], df['GDP_Growth'], label = 'GDP_Growth')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.subplot(3, 1, 2)
plt.plot(df['Year'], df['Inflation'], label = 'Inflation' )
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.subplot(3, 1, 3)
plt.plot(df['Year'], df['Unemployment'], label = 'Unemployment')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.grid()
```

```python
plt.subplot(1, 3, 1)
plt.plot(df['Year'], df['GDP_Growth'], label = 'GDP_Growth')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.subplot(1, 3, 2)
plt.plot(df['Year'], df['Inflation'], label = 'Inflation' )
plt.legend()
plt.xlabel('Year')
plt.subplot(1, 3, 3)
plt.plot(df['Year'], df['Unemployment'], label = 'Unemployment')
plt.legend()
plt.xlabel('Year')
plt.grid()
```
