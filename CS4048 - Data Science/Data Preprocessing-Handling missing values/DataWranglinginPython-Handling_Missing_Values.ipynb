Filling Missing Values

```python
df[' BMI '].skew()
```

```python
bmiM=df[' BMI '].mean()
df[' BMI '].fillna(bmiM,inplace=True)
```

```python
df.isnull().sum()
```

Fill Using Ffill/Bfil

```python
df_ffill=df.ffill()
df_ffill.isnull().sum()
```

Fill Using Interpolation

```python
df_inter=df.select_dtypes(include=['float64','int64']).interpolate(method='linear')
df_inter.isnull().sum()
```

```python
concatdf=pd.concat([df_inter,df.select_dtypes(exclude=['float64', 'int64'])],axis=1)
concatdf.isnull().sum()
```
