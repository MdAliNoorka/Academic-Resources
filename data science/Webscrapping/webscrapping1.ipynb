```python
faculty=soup.find('div',id='fsc').findAll('div', class_='card')
```

```python
faculty[0]
```

```python
NU='http://lhr.nu.edu.pk/'
```

```python
ID=[]
FacultyURL=[]
email=[]
name=[]
image=[]
designation=[]

```

```python
for f in faculty:
    ID.append(f.find('a',class_='faculty-link')['href'].split('/')[-1])
    FacultyURL.append(NU+f.find('a',class_='faculty-link')['href'])
    email.append(f.find('p', class_='mb-0').text.strip())
    name.append(f.find('h5',class_='text-center').text.strip())
    designation.append(f.find('p', class_='small').text.strip().split('\n')[0])
    image.append(NU+f.find('img', class_='card-img-top')['src'])
    
```

```python
fsc=pd.DataFrame({'ID':ID,
                  'name':name,
                  'email':email,
                  'designation':designation,
                  'FacultyURL': FacultyURL,
                  'image':image})
```

```python
lecturer=fsc[fsc.designation=='Lecturer']
len(lecturer)
```

```python
fsc
```

```python
import os

# Create a directory for saving images
os.makedirs("faculty_images", exist_ok=True)

# Download and save images
for i, img_url in enumerate(lecturer['image']):
    img_data = requests.get(img_url).content
    with open(f"faculty_images/image_{i+1}.jpg", "wb") as file:
        file.write(img_data)
    print(f"Downloaded: {img_url}")
```

```python
counter=0
for i in lecturer['image']:
    re=requests.get(i);
    file_name="faculty_images/"+"image_"+str(counter)+".jpg"
    with open(file_name,"wb") as file:
        file.write(re.content)
    counter+=1
    
```

```python
# The below url contains an html table with data about colors and color codes.
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
data = requests.get(url).text 
```

```python
soup = BeautifulSoup(data, "html.parser")
table = soup.find('table')  # in html table is represented by the tag <table>
```

```python
color_name=[]
color_code=[]
# Get all rows from the table
for row in table.find_all('tr')[1:]:  # in html table row represented by tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td')  # in html a column is represented by tag <td>
    color_name.append(cols[2].text)  # store the value in column 3 as color_name
    color_code.append(cols[3].text ) # store the value in column 4 as color_code
    print("{}--->{}".format(color_name[-1], color_code[-1]))
```

```python

tabledata=pd.DataFrame({'color_name':color_name,
                  'color_code':color_code
                  })


tabledata.head(100)
```
