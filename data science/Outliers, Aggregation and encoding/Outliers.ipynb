```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame
# Calculate skewness for numeric columns
skewness = df.select_dtypes(include=['float64', 'int64']).skew()

# Number of columns
num_columns = len(skewness)

# Calculate the number of rows needed for 4 graphs per row
num_rows = (num_columns + 3) // 4  # Add 3 to ensure rounding up

# Create subplots with 4 graphs per row
fig, axes = plt.subplots(nrows=num_rows, ncols=4, figsize=(20, 5 * num_rows))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot skewness for each column
for i, (col, skew_val) in enumerate(skewness.items()):
    sns.histplot(df[col], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'{col}\nSkewness: {skew_val:.2f}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Selecting only the required columns

df_selected = df[['Alcohol', 'Total expenditure']].copy()

# Fill missing values using backfill
df_selected.bfill(inplace=True)

# Sorting values for each column separately
for col in df_selected.columns:
    df_selected[col] = df_selected[col].sort_values().values

# Calculating quartiles, IQR, and outlier bounds for selected columns
stats = {}
for column in df_selected.columns:
    Q1 = np.percentile(df_selected[column], 25)  # First quartile (25%)
    Q3 = np.percentile(df_selected[column], 75)  # Third quartile (75%)
    median = np.median(df_selected[column])  # Median (Q2)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - (1.5 * IQR)  # Lower bound
    upper_bound = Q3 + (1.5 * IQR)  # Upper bound
    min_val = df_selected[column].min()  # Minimum value
    max_val = df_selected[column].max()  # Maximum value

    stats[column] = {
        "Q1 (25%)": Q1,
        "Q3 (75%)": Q3,
        "Median (Q2)": median,
        "IQR": IQR,
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Min Value": min_val,
        "Max Value": max_val,
    }

# Convert stats to a DataFrame for display
stats_df = pd.DataFrame(stats)

# Creating the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_selected, width=0.5, palette="coolwarm")



plt.title("Boxplot with Quartiles and Boundaries for Alcohol & Total Expenditure")
plt.show()

# Display the statistics DataFrame
print(stats_df)
```

```python
Q1 = df['Life expectancy '].quantile(0.25)
Q3 = df['Life expectancy '].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecting outliers
outliers = df[(df['Life expectancy '] < lower_bound) | (df['Life expectancy '] > upper_bound)]

outliers
```

```python
from scipy import stats
# Calculate Z-scores
non_null_filter = df['Life expectancy '].notnull()
#print(non_null_filter)

df_filtered = df[non_null_filter]

life_expectancy_values = df_filtered['Life expectancy ']
df['Z_score'] = stats.zscore(life_expectancy_values)

# Set threshold for outliers
outliers = df[df['Z_score'].abs() > 3]

outliers
```
